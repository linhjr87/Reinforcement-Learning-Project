# Summary: Four Research Directions on Vasuki

This notebook does **no training** and embeds **no environment** — it only reads the JSON
result files saved by the four direction notebooks (`direction1_selfplay_league.ipynb` ...
`direction4_pomdp_elo.ipynb`) from `results/` and aggregates them into a single master table
and an Elo ranking chart.

Run notebooks 1 -> 4 first (each writes its own `results/direction<N>_*.json`, and direction 4
additionally writes `results/elo_ratings.json`), then run this notebook last.


```python
import json
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

rows = []
for path in sorted(Path("results").glob("direction*.json")):
    for rec in json.loads(path.read_text()):
        rows.append({"file": path.name, "direction": rec["direction"],
                     "experiment": rec["experiment"], "seed": rec["seed"],
                     **{f"cfg_{k}": str(v) for k, v in rec["config"].items()},
                     **{f"m_{k}": (v if not isinstance(v, dict) else json.dumps(v))
                        for k, v in rec["metrics"].items()}})
if not rows:
    print("WARNING: no results found — run the direction notebooks first.")
df = pd.DataFrame(rows)
df
```

```python
elo_path = Path("results/elo_ratings.json")
if elo_path.exists():
    elo = json.loads(elo_path.read_text())
    names = sorted(elo["ratings"], key=elo["ratings"].get)
    vals = [elo["ratings"][n] for n in names]
    errs = [[elo["ratings"][n] - elo["ci95"][n][0] for n in names],
            [elo["ci95"][n][1] - elo["ratings"][n] for n in names]]
    plt.barh(names, vals, xerr=errs)
    plt.xlabel("Elo (K=32, 95% bootstrap CI)"); plt.title("Agent ranking"); plt.show()
else:
    print("WARNING: results/elo_ratings.json missing — run direction 4 first.")
```
